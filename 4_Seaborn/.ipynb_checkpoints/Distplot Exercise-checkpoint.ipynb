### 1) Import the rrequired dependencies to draw seaborn plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2) Load the mpg dataset and store it in a variable called cars


```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

### 3) Draw a distplot for the displacement column of the mpg dataset

```python
sns.distplot(df['displacement'])
```

### 4) Disable the KDE Curve from the above plot using required the required function

```python
sns.distplot(df['displacement'], kde=False)
```

### 5) Enable the rugplot in the above plot using required function

```python
sns.distplot(df['displacement'], rug_kws={'color':'red', 'height':0.30}, rug=True)
```

### 6) Set the number of bins to 20 in above plot


```python
sns.distplot(df['displacement'], bins=20,  rug_kws={'color':'red', 'height':0.30}, rug=True)
```

### Convert the plot above to a normal distribution with no of bins as 10 8.Change the transperency of the histogram in the above plot to 0.6 and also the hist type to step

```python
sns.distplot(df['displacement'], bins=10, hist_kws={'color':'pink', 'alpha':0.6, 'histtype':'step'})
```

### Change the color of the above plot to red


```python
sns.distplot(df['displacement'], bins=10, hist_kws={'color':'red', 'alpha':0.6, 'histtype':'step'})
```

### Draw a distplot with the displacement as the column and change the line width of the KDE Plot to 5 11.In the above plot change the color of the rugplot to red and the height


```python
sns.distplot(df['displacement'], bins=20, rug_kws={'color':'red', 'height':5.11}, rug=True)
```
