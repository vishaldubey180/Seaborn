### Distplot

### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Loading dataset

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

### Drawing simple distplot

```python
sns.distplot(df['flipper_length_mm'], bins=1)
```

```python
sns.distplot(df.flipper_length_mm, bins=1, rug=True)
```

#### Removing KDE plot

```python
sns.distplot(df.flipper_length_mm, bins=1, kde=False)
```

```python
sns.distplot(df.flipper_length_mm, bins=1, hist=False)
```

### setting the number of bins

```python
sns.distplot(df['flipper_length_mm'], bins=20)
```

### Setting the bin values as an array

```python
sns.distplot(df['flipper_length_mm'], bins=[160,170,180,190,200])
```

```python
sns.distplot(df['flipper_length_mm'], bins=[160,170,180,190,200,240])
```

### Fitting the data to particular distribution using norm_list

```python
sns.distplot(df['flipper_length_mm'], bins=20, norm_hist=True)
```

```python
sns.distplot(df['flipper_length_mm'], bins=20, norm_hist=True, kde=False)
```

### Styling Rugplots

```python
sns.distplot(df['flipper_length_mm'], bins=20, rug_kws={'color':'red', 'height':0.30}, rug=True)
```

```python
sns.distplot(df['flipper_length_mm'], bins=20, hist_kws={'color':'pink', 'alpha':1, 'histtype':'step'})
```

### Styling KDE Plot

```python
sns.distplot(df['flipper_length_mm'], bins=20, kde=True, hist=False, kde_kws={'color':'red', 'alpha':1, 'lw':2.5})
```
